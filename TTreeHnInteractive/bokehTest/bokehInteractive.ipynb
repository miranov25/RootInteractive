```python
import numpy as np # we will use this later, so import it now
from bokeh.layouts import gridplot
from bokeh.models import ColumnDataSource, CDSView, BooleanFilter
from bokeh.plotting import figure, show
from bokeh.io import output_notebook, show
from ipywidgets import *
from IPython.display import display
output_notebook()
```

```python
n = 1000
x = np.random.standard_normal(n)
y = np.random.standard_normal(n)
z = np.random.standard_normal(n)
data = {'x': x,'y': y,'z':z}
source = ColumnDataSource(data=data)
tools = ["box_select", "hover", "reset"]
p = figure(plot_height=500, plot_width=500, tools=tools)
p.circle(x="x", y="y", size=3, hover_color="red", source=source)
#show(p)
booleansx=0
booleansy=0
p_filtered=0
plot=0;
#show(p_filtered)
def update(cutx=0,cuty=0):
    global booleansx,boolensy, p_filtered,plot
    p_filtered = figure(plot_height=500, plot_width=500, tools=tools, x_range=(-3, 3), y_range=(-3, 3))
    booleansy = [True if y_val > cuty else False for y_val in source.data['y']]
    booleansx = [True if x_val > cutx else False for x_val in source.data['x']]
    view = CDSView(source=source, filters=[BooleanFilter(booleansx), BooleanFilter(booleansy)])
    #p_filtered = figure(plot_height=500, plot_width=500, tools=tools, x_range=p.x_range, y_range=p.y_range)
    plot=p_filtered.circle(x="x", y="y", size=3, hover_color="red", source=source, view=view)
    #p.view=view
    show(p_filtered)
    
interact(update,cutx=(-3,3,0.1), cuty=(-3,3,0.1))    

```
